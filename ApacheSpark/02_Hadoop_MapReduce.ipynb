# Configurar Hadoop de Nodo Único

Los pasos descritos en este laboratorio utilizan la versión 3.3.6 de Hadoop. **Hadoop** es más útil cuando se despliega en un modo completamente distribuido en un gran clúster de servidores en red que comparten un gran volumen de datos. Sin embargo, para una comprensión básica, configuraremos Hadoop en un solo nodo.

En este laboratorio, ejecutaremos el ejemplo WordCount con un texto de entrada y veremos cómo se procesa el contenido del archivo de entrada por WordCount.

1. Inicie un nuevo terminal

    ![](https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-BD0225EN-SkillsNetwork/labs/images/New_terminal.png)

2. Descargue hadoop-3.2.3.tar.gz a su entorno theia ejecutando el siguiente comando.

```
curl https://dlcdn.apache.org/hadoop/common/hadoop-3.3.6/hadoop-3.3.6.tar.gz --output hadoop-3.3.6.tar.gz
```

3. Extrae el archivo tar en el directorio actual.

```
tar -xvf hadoop-3.3.6.tar.gz
```

4. Navega al directorio hadoop-3.3.6.

```
cd hadoop-3.3.6
```

5. Verifica el comando hadoop para ver si está configurado. Esto mostrará la documentación de uso para el script hadoop.

```
bin/hadoop
```

6. Ejecuta el siguiente comando para descargar data.txt en tu directorio actual.

```
curl https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-BD0225EN-SkillsNetwork/labs/data/data.txt --output data.txt
```

7. Ejecuta la aplicación Map reduce para el conteo de palabras en data.txt y almacena la salida en **/user/root/output**

```
bin/hadoop jar share/hadoop/mapreduce/hadoop-mapreduce-examples-3.3.6.jar wordcount data.txt output
```

> Esto puede tardar un poco.

8. Una vez que el conteo de palabras se ejecute correctamente, puedes ejecutar el siguiente comando para ver el archivo de salida que ha generado.

```
ls output
```

Deberías ver **part-r-00000** con **\_SUCCESS** indicando que se ha completado el conteo de palabras.

> Mientras aún se esté procesando, es posible que solo veas _‘\_temporary’_ listado en el directorio de salida. Espera un par de minutos y ejecuta el comando nuevamente hasta que veas la salida como se muestra arriba.

9. Ejecuta el siguiente comando para ver la salida del conteo de palabras.

```
cat  output/part-r-00000
```
![Salida local de wordcount](https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-BD0225EN-SkillsNetwork/labs/images/wordcount_output.png)
